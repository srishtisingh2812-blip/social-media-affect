```
import pandas as pd

df = pd.read_csv("Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")
df.head()
```

```
df.info()
df.describe()
```

```
df.isnull().sum()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['age'], kde=True)
plt.title("age Distribution")
plt.show()
```

```
sns.countplot(x='gender', data=df)
plt.title("gender Distribution")
plt.show()
```

```
sns.scatterplot(x='daily_screen_time_hours', y='addicted_label', data=df)
plt.title("Screen Time vs Addiction")
plt.show()
```

```
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```
sns.boxplot(x='gender', y='daily_screen_time_hours', data=df)
plt.title("Screen Time by Gender")
plt.show()
```

```
sns.barplot(x='gender', y='daily_screen_time_hours', data=df)
plt.title("Average Screen Time by Gender")
plt.show()
```
